```python
def plotMosaico(img, titulos, alt, anch, nFilas, nColumnas, random):
    """
    Muestra una galería de nFilas x nColumnas con las imágenes contenidas en el arreglo img
    Asigna a cada imagen uno de los títulos contenidos en el arreglo titulos, cada imagen es de alt x anch
    El parámetro random al estar activado selecciona imágenes al azar del arreglo img
    """
    plt.figure(figsize=(1.7 * nColumnas, 2.3 * nFilas))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(nFilas * nColumnas):
        plt.subplot(nFilas, nColumnas, i + 1)
        if random:
            random = np.random.randint(0, img.shape[0])
            plt.imshow(img[random].reshape((alt, anch)), cmap=plt.cm.gray)
            if titulos != np.array([]):
                plt.title(titulos[random], size=12)
        else:
            plt.imshow(img[i].reshape((alt, anch)), cmap=plt.cm.gray)
            if titulos != np.array([]):
                plt.title(titulos[i], size=12)
        plt.xticks(())
        plt.yticks(())
```

```python
def muestraImagen(img, alt, anch, titulo):
    """
    Muestra la imagen img, con las medidas alt x anch; le asigna el título dado
    """
    plt.imshow(img.reshape((alt, anch)), cmap=plt.cm.gray)
    plt.title(titulo)
    plt.xticks(())
    plt.yticks(())
```

```python
def minimoImagenes():
    """
    Devuelve el menor número de imágenes que tienen los personajes
    """
    mi = (claves == 0).sum()
    for i in range(0,nPersonas):
        if (claves == i).sum() < mi:
            mi = (claves == i).sum()
    return mi
```

```python
#Cantidad de imágenes de entrenamiento
nEntrenamiento = minimoImagenes() - 1
```

```python
def entrenamientoPrueba(clave):
    """
    Separa en dos grupos las imágenes de un personaje.
    Devuelve dos arreglos, uno con las imágenes de entrenamiento y otro con las de prueba
    """
    m = minimoImagenes()
    res = (claves == clave).sum() - (nEntrenamiento + 1)
    imgPer = imgPersona(clave)
    entre = imgPer[0:nEntrenamiento]
    prueba = imgPer[nEntrenamiento:]
    return entre, prueba
```

Guardamos las imágenes de entrenamiento y de prueba en dos arreglos llamados `entrenamientos` y `pruebas` donde cada entrada guarda las imágenes de entrenameinto o prueba correspondientes a una sola persona

```python
def caricaturasPersona(clave):
    """
    Recibe un arreglo con imágenes y devuelve un arreglo con las caricaturas del personaje.
    La caricatura se obtiene restando el promedio de una persona a una imagen.
    """
    entr = entrenamientoPrueba(clave)[0]
    #prom = promedios[clave]
    car = np.array(np.arange(entr.shape[0]), dtype = object)
    for i in range(0, entr.shape[0]):
        car[i] = entr[i] - promedioGeneral
    return car

def arrToMatrix(arrEntre):
    """
    Recibe un arreglo de una dimensión con las imágenes de entrenamiento y lo transforma en una matriz de (no_imag, 2914)
    """
    arr = arrEntre
    s = arr.shape[0]
    resp = np.zeros(62*47*s)
    for i in range(0, s):
        arr[i] = arr[i].reshape(62*47,1)
    for i in range(0,s):
        for j in range(0, 62*47):
            resp[i*62*47 + j] = arr[i][j]
    return np.asmatrix(resp.reshape(s, 62*47))
```

<b>Proyección</b>
</div>

Proyectar una imagen sobre el espacio generado por los eigenvectores 'cara espacio' nos permite saber si la imagen se encuentra cerca de este espacio. Si se encuentra cerca podemos intuir que se trata de una imagen de una persona.

```python
def norma(vec):
    """
    Calcula la norma de un vector o arreglo de una dimensión
    """
    v = np.asarray(vec)
    norma = ((v**2).sum())**0.5
    return norma

def proyeccion(imagen, base):
    """
    Recibe una imagen y un arreglo de eigenvectores.
    Devuelve un arreglo con la proyección de la imagen sobre el espacio generado por los eigenvectores.
    """
    #proy = np.asmatrix(np.zeros(62*47).reshape(62*47,1))
    b = np.asmatrix(imagen.reshape(62*47,1) - promedioGeneral.reshape(62*47,1))
    U = np.asmatrix(base)
    proy = U*U.T*b
    return proy

def coeficientes(imagen, base):
    """
    Dada una imagen y una base, devuelve un arreglo con los coeficientes de la combinación lineal de la imagen en esa base
    """
    b = np.asmatrix(imagen.reshape(62*47,1) - promedioGeneral.reshape(62*47,1))
    U = np.asmatrix(base)
    coefs = U*b
    return np.asarray(coefs)
```

Nuestro primer método consiste en comparar la imagen a clasificar con cada una de las firmas

```python
def diferenciaFirma(imagen, clave):
    """
    Recibe una imagen y la compara con la firma de una clave determinada
    """
    im = coeficientes(imagen, eigVecMRed).T
    return norma(im - firmas[clave])

def prediccion(imagen):
    minimo = diferenciaFirma(imagen, 0)
    p = 0
    for i in range (0, nPersonas):
        d = diferenciaFirma(imagen, i)
        #print str(i) + "\t" + str(d) + "\t" + str(minimo)
        if (d - minimo) < 0:
            minimo = d
            p = i
            #print str(indice) + "\t" + str(minimo)
    return p

def porAciertoIndividual(clave):
    aci = 0
    for j in range(0, pruebas[clave].shape[0]):
        #print "Otra imagen"
        if prediccion(pruebas[clave][j]) == clave:
            aci = aci + 1;
    return 100.0*aci/pruebas[clave].shape[0]
```

Nuestro segundo método consiste en comparar la imagen a clasificar con cada una de las firmas y verificar si la diferencia se encuentra por debajo de un parámetro determinado

```python
def prediccion2(imagen, parametro):
    minimo = diferenciaFirma(imagen, 0)
    p = -1
    for i in range (0, nPersonas):
        d = diferenciaFirma(imagen, i)
        if d < parametro:
            if (d - minimo) < 0:
                minimo = d
                p = i
    return p

def porAciertoIndividual2(clave, parametro):
    aci = 0
    desc = 0
    for j in range(0, pruebas[clave].shape[0]):
        pred = prediccion2(pruebas[clave][j], parametro)
        if pred == -1:
            desc = desc + 1
        else:
            if pred == clave:
                aci = aci + 1;
    if desc == pruebas[clave].shape[0]:
        resp = -1
    else:
        resp = 100.0*aci/(pruebas[clave].shape[0] - desc)
    return resp
```

Nuestro tercer método consiste en comparar a la imagen con cada uno de las que están en el conjunto de entrenamiento

```python
def diferenciaEntrenamiento(imagen, clave):
    """
    Recibe una imagen y la compara con cada uno de los elementos del conjunto de entrenamiento de una clave determinada
    """
    resp = np.zeros(nEntrenamiento)
    img = imagen.reshape(2914,1)
    for i in range(0, nEntrenamiento):
        entr = entrenamientos[clave][i].reshape(2914,1)
        numr = norma(entr - img)
        denm = norma(entr)
        resp[i] = numr/denm
    return resp

def predA(imagen):
    minimo16a = norma(diferenciaEntrenamiento(imagen, 0))
    pred16a = 0
    for i in range(0, nPersonas):
        n = norma(diferenciaEntrenamiento(imagen, i))
        if n < minimo16a:
            minimo16a = n;
            pred16a = i;
    return pred16a;

def porAciertoIndividualA(clave):
    aci = 0
    for j in range(0, pruebas[clave].shape[0]):
        if predA(pruebas[clave][j]) == clave:
            aci = aci + 1;
    return 100.0*aci/pruebas[clave].shape[0]
```

```python
def diferenciaEJProy(imagen, clave):
    """
    Recibe una imagen y compara su proyección con cada una de las proyecciones del conjunto de entrenamiento de una clave determinada
    """
    resp = np.zeros(no_eigC)
    coefsImg = coeficientes(imagen.reshape(2914,1), basesNormalizadas[clave],no_eigC)
    for i in range(0, no_eigC):
        coefsEntr = coeficientes(entrenamientos[clave][i].reshape(2914,1), basesNormalizadas[clave],no_eigC)
        numr = np.asarray(coefsEntr - coefsImg)
        numr = norma(numr)
        denm = np.asarray(coefsEntr)
        denm = norma(denm)
        resp[i] = numr/denm
    return resp
```
