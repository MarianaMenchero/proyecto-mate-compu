```python
def plotMosaico(img, titulos, alt, anch, nFilas, nColumnas, random):
    """
    Muestra una galería de nFilas x nColumnas con las imágenes contenidas en el arreglo img
    Asigna a cada imagen uno de los títulos contenidos en el arreglo titulos, cada imagen es de alt x anch
    El parámetro random al estar activado selecciona imágenes al azar del arreglo img
    """
    plt.figure(figsize=(1.7 * nColumnas, 2.3 * nFilas))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(nFilas * nColumnas):
        plt.subplot(nFilas, nColumnas, i + 1)
        if random:
            random = np.random.randint(0, img.shape[0])
            plt.imshow(img[random].reshape((alt, anch)), cmap=plt.cm.gray)
            if titulos != np.array([]):
                plt.title(titulos[random], size=12)
        else:
            plt.imshow(img[i].reshape((alt, anch)), cmap=plt.cm.gray)
            if titulos != np.array([]):
                plt.title(titulos[i], size=12)
        plt.xticks(())
        plt.yticks(())
```

### <div style = "background-color: #C0E159; border-radius: 3px; padding: 7px"> <b>Rostro promedio por persona</b> </div>

Cada imagen es un arreglo de $1 \times 2914$, podemos obtener la cara promedio de una persona dada su clave sacando el promedio de las imágenes de la siguiente forma:

$$\bar{\phi} = \frac{1}{M}\sum_{i = 1}^{M}\phi^{(i)}$$

donde $\phi^{(i)}$ es la i-ésima imagen de la persona

```python
def rostroPromTotal(clave):
    """
    Regresa el rostro promedio de una persona dada su clave
    """
    cant = cuantasFotos(clave, claves)
    prom = np.zeros((62,47))
    for i in range(0, cant):
        prom = prom + imgPersona(clave)[i]
    prom = prom / cant
    return prom
```

```python
def entrenamientoPrueba(clave):
    """
    Separa en dos grupos las imágenes de un personaje.
    Devuelve dos arreglos, uno con las imágenes de entrenamiento y otro con las de prueba
    """
    m = minimoImagenes()
    res = (claves == clave).sum() - (nEntrenamiento + 1)
    imgPer = imgPersona(clave)
    entre = imgPer[0:nEntrenamiento]
    prueba = imgPer[nEntrenamiento:]
    return entre, prueba
```

### <div style = "background-color: #C0E159; border-radius: 3px; padding: 7px"> <b>Matriz M (imágenes de entrenamiento)</b> </div>

La caricatura de la i-ésima imagen se define como: $\Phi^{(i)} = \phi^{(i)} - \bar{\phi}$ donde $\bar{\phi}$ es el promedio general del conjunto de entrenamiento.

Creamos una matriz $M$ de $76\cdot6 \times 2914$ con las caricaturas del conjunto de entrenamiento

```python
def caricaturasPersona(clave):
    """
    Recibe un arreglo con imágenes y devuelve un arreglo con las caricaturas del personaje.
    La caricatura se obtiene restando el promedio de una persona a una imagen.
    """
    entr = entrenamientoPrueba(clave)[0]
    car = np.array(np.arange(entr.shape[0]), dtype = object)
    for i in range(0, entr.shape[0]):
        car[i] = entr[i] - promedioGeneral
    return car

def arrToMatrix(arrEntre):
    """
    Recibe un arreglo de una dimensión con las imágenes de entrenamiento y lo transforma en una matriz de (no_imag, 2914)
    """
    arr = arrEntre
    s = arr.shape[0]
    resp = np.zeros(62*47*s)
    for i in range(0, s):
        arr[i] = arr[i].reshape(62*47,1)
    for i in range(0,s):
        for j in range(0, 62*47):
            resp[i*62*47 + j] = arr[i][j]
    return np.asmatrix(resp.reshape(s, 62*47))
```

Creamos un arreglo con las caricaturas del conjunto de entrenamiento por persona

```python
arrCaricaturas = np.array(np.arange(nPersonas), dtype=object)
for i in range(0,nPersonas):
    arrCaricaturas[i] = caricaturasPersona(i)
```

```python
#Esta matriz tiene
#matrizA = np.array(np.arange(nPersonas), dtype=object)
#for i in range(0,nPersonas):
#    matrizM[i] = arrToMatrix(arrCaricaturas[i])
```

```python
matrizM = np.zeros((nEntrenamiento*nPersonas,2914))
for i in range(0,nPersonas):
    for j in range(0,nEntrenamiento):
        matrizM[i*nEntrenamiento + j] = arrCaricaturas[i][j].reshape(1,2914)
matrizM = np.asmatrix(matrizM)
```

Definimos la matriz de covarianza de $M$ como:

$$C = M^TM = \sum_{i = 1}^{m}\Phi^{(i)T}\Phi^{(i)}$$

donde $m$ es la cantidad de imágenes en el conjunto de entrenamiento. La matriz $C$ tiene una dimensión de $2914 \times 2914$

<div style = "background-color: #C0E159; border-radius: 3px; padding: 7px">
<b>Eigenvalores y eigenvectores</b>
</div>

Como la matriz de covarianza ($C = M^T M$) es muy grande es computacionalmente costoso e impreciso calcular sus eigenvalores y eigenvectores, por lo que consideraremos la matriz $MM^T$ de dimensión $76\cdot6 \times 76\cdot6$. Esto es posible por lo siguiente:

$$ M M^T v_i = \mu_i v_i \Rightarrow M^T M M^T v_i = \mu_i M^T v_i$$

$$\Rightarrow CM^T v_i = \mu_i M^T v_i \Rightarrow C \omega_i = \mu_i \omega_i $$ donde $\omega_i = M^T v_i$

```python
matrizChica = matrizM * matrizM.T
```

Calculamos los eigenvalores y eigenvectores de esta nueva matriz

```python
eigValM, eigVecCh = eig(matrizChica)
```

Obtenemos los eigenvectores de la matriz de covarianza ($C$)

```python
eigVecM = matrizM.T * eigVecCh
```

Elegimos los eigenvectores con los mayores eigenvalores asociados porque son los que aportan más información.

```python
def indicesMayores(arr, n):
    """
    Regresa los índices donde se encuentran los n mayores valores de un arreglo
    """
    ind = arr.argsort()[-n:][::-1]
    return ind
```

```python
#Cantidad de eigencaras
no_eigC = 30
```

```python
indicesMay = indicesMayores(eigValM,no_eigC)
```

Creamos nuevos arreglos con los eigenvalores y los eigenvectores que elegimos

```python
eigVecMRed = eigVecM.T[indicesMay]
eigValMRed = eigValM[indicesMay]
```

Para poder proyectar sobre el espacio de eigencaras, normalizamos la base de eigenvectores

```python
def norma(vec):
    """
    Calcula la norma de un vector o arreglo de una dimensión
    """
    v = np.asarray(vec)
    norma = ((v**2).sum())**0.5
    return norma
```

```python
for i in range(0, no_eigC):
    eigVecMRed[i] = eigVecMRed[i] / norma(eigVecMRed[i])
```

### <div style = "background-color: #C0E159; border-radius: 3px; padding: 7px"> <b>Proyección</b> </div>

Proyectar una imagen sobre el espacio generado por los eigenvectores 'cara espacio' nos permite saber si la imagen se encuentra cerca de este espacio. Si se encuentra cerca podemos intuir que se trata de una imagen de una persona.

```python
def proyeccion(imagen, base):
    """
    Recibe una imagen y un arreglo de eigenvectores.
    Devuelve un arreglo con la proyección de la imagen sobre el espacio generado por los eigenvectores.
    """
    b = np.asmatrix(imagen.reshape(62*47,1) - promedioGeneral.reshape(62*47,1))
    U = np.asmatrix(base)
    proy = U*U.T*b
    return proy

def coeficientes(imagen, base):
    """
    Dada una imagen y una base, devuelve un arreglo con los coeficientes de la combinación lineal de la imagen en esa base
    """
    b = np.asmatrix(imagen.reshape(62*47,1) - promedioGeneral.reshape(62*47,1))
    U = np.asmatrix(base)
    coefs = U*b
    return np.asarray(coefs)
```

Guardamos los coeficientes de las proyecciones de cada imagen de entrenamiento en un arreglo

```python
coefs = np.array(np.arange(nPersonas), dtype = object)
for i in range(0, nPersonas):
    pr = cuantasFotos(i) - nEntrenamiento
    coefs[i] = np.array(np.arange(pr), dtype = object)
    for j in range(0, pr):
        coefs[i][j] = coeficientes(pruebas[i][j], eigVecMRed)
```

Creamos un vector con las proyecciones de los promedios de cada persona (firmas)

```python
firmas = np.zeros((nPersonas, no_eigC))
for i in range(0,nPersonas):
    firmas[i] = coeficientes(promedios[i],eigVecMRed).T[0]
```

### <div style = "background-color: #C0E159; border-radius: 3px; padding: 7px"> <b>Clasificación de rostros</b> </div>

Nuestro primer método consiste en comparar la imagen a clasificar con cada una de las firmas

```python
def diferenciaFirma(imagen, clave):
    """
    Recibe una imagen y la compara con la firma de una clave determinada
    """
    im = coeficientes(imagen, eigVecMRed).T
    return norma(im - firmas[clave])

def prediccion(imagen):
    minimo = diferenciaFirma(imagen, 0)
    p = 0
    for i in range (0, nPersonas):
        d = diferenciaFirma(imagen, i)
        if (d - minimo) < 0:
            minimo = d
            p = i
    return p

def porAciertoIndividual(clave, conjunto):
    aci = 0
    resp = 0
    if (conjunto == "prueba"):
        for j in range(0, pruebas[clave].shape[0]):
            if prediccion(pruebas[clave][j]) == clave:
                aci = aci + 1;
        resp = 100.0*aci/pruebas[clave].shape[0]
    else:
        if (conjunto == "entrenamiento"):
            for j in range(0, entrenamientos[clave].shape[0]):
                if prediccion(entrenamientos[clave][j]) == clave:
                    aci = aci + 1;
            resp = 100.0*aci/entrenamientos[clave].shape[0]
    return resp

def porAciertoIndividualEntrenamiento(clave):
    aci = 0
    for j in range(0, entrenamientos[clave].shape[0]):
        if prediccion(entrenamientos[clave][j]) == clave:
            aci = aci + 1;
    return 100.0*aci/entrenamientos[clave].shape[0]
```

Nuestro segundo método consiste en comparar la imagen a clasificar con cada una de las firmas y verificar si la diferencia se encuentra por debajo de un parámetro determinado

```python
def prediccion2(imagen, parametro):
    minimo = diferenciaFirma(imagen, 0)
    p = -1
    for i in range (0, nPersonas):
        d = diferenciaFirma(imagen, i)
        if d < parametro:
            if (d - minimo) < 0:
                minimo = d
                p = i
    return p

def porAciertoIndividual2(clave, parametro, conjunto):
    aci = 0
    desc = 0
    if (conjunto == "prueba"):
        for j in range(0, pruebas[clave].shape[0]):
            pred = prediccion2(pruebas[clave][j], parametro)
            if pred == -1:
                desc = desc + 1
            else:
                if pred == clave:
                    aci = aci + 1;
        if desc == pruebas[clave].shape[0]:
            resp = -1
        else:
            resp = 100.0*aci/(pruebas[clave].shape[0] - desc)
    else:
        if conjunto == "entrenamiento":
            for j in range(0, entrenamientos[clave].shape[0]):
                pred = prediccion2(entrenamientos[clave][j], parametro)
                if pred == -1:
                    desc = desc + 1
                else:
                    if pred == clave:
                        aci = aci + 1;
            if desc == entrenamientos[clave].shape[0]:
                resp = -1
            else:
                resp = 100.0*aci/(entrenamientos[clave].shape[0] - desc)
    return resp
```
